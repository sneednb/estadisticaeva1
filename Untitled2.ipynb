```python
# Importar librería
import pandas as pd
muestra = df_1.shape[0]
muestra



intervalos = 10

# Vamos a agregar una nueva columna, con la función cut()
df_1['Intervalos'] = pd.cut(df_1['VALOR ARANCEL (PESOS)'], bins = intervalos,
                                  precision = 1, include_lowest=True)

# Frecuencias
f = df_1.groupby('Intervalos').size()
F = f.cumsum()
h = (f/muestra)*100
H = h.cumsum()

# Armar el DataFrame
tabla6 = pd.DataFrame({
    'Frecuencia absoluta': f,
    'Frecuencia relativa (%)': h,
    'Frecuencia absoluta acumulada': F,
    'Frecuencia relativa acumulada (%)': H
})

tabla6
```

# **PREGUNTA 1**
¿HAY ÁREAS DEL CONOCIMIENTO DONDE LAS CARRERAS SEAN MÁS CARAS? EXPLIQUEN QUE
CRITERIO UTILIZARON PARA RESPONDER A ESTA PREGUNTA.
En esta pregunta deben crear una tabla dinámica utilizando las variables que corresponden
para abordarla. En esta pregunta resultan fundamentales 3 puntos:
1) Debe analizar qué indicadores estadísticos de los vistos en la unidad 1 son útiles y
necesarios para abordar la pregunta y, por ende, son apropiados para ser
mostrados en la tabla dinámica.
2) Creación correcta de la tabla dinámica y gráfico apropiado.
3) Análisis y conclusión en torno a la información obtenida.

Los valores estadisticos que nos interesan para responder la pregunta son los de tendencia central y dispersión.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#CREACIÓN DE LA TABLA DINÁMICA---


tabla_dispersion = df_1.groupby('AREA CONOCIMIENTO')['VALOR ARANCEL (PESOS)'].agg(
    Media_Arancel = 'mean',
    Maximo_Arancel = 'max',
    Desviacion_Estandar = 'std',
    N_Carreras = 'count'
).reset_index()

#cv
tabla_dispersion['Coeficiente_Variacion (%)'] = (
    tabla_dispersion['Desviacion_Estandar'] / tabla_dispersion['Media_Arancel']
) * 100

#ordenar cv descendente
#cv mas alto mayor disparidad de valores
tabla_dispersion_ordenada = tabla_dispersion.sort_values(
    by='Coeficiente_Variacion (%)',
    ascending=False
)

#tabla actualizada
tabla_dispersion_ordenada.columns = [
    'Área del Conocimiento',
    'Media',
    'Máximo',
    'Desviación Estándar',
    'N° Carreras',
    'CV (%)'
]

print("--- TABLA DINÁMICA: Dispersión de Aranceles (Ordenada por CV) ---")
print(tabla_dispersion_ordenada.round(2))

# ---grafico de c.v--
#cv indicador principal
plt.figure(figsize=(12, 6))

plt.bar(
    tabla_dispersion_ordenada['Área del Conocimiento'],
    tabla_dispersion_ordenada['CV (%)'],
    color='purple',
    edgecolor='black'
)

#titulos
plt.title('Coeficiente de Variación (CV) del Arancel por Área del Conocimiento', fontsize=14)
plt.xlabel('Área del Conocimiento', fontsize=12)
plt.ylabel('Coeficiente de Variación (%)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

# **PREGUNTA 3**
¿HAY CARRERAS CUYO ARANCEL SEA SUSTANTIVAMENTE MÁS CARO QUE LA MAYORÍA DE LOS
ARANCELES? ¿QUÉ EXPLICACIÓN PUEDEN ENCONTRAR?
En esta pregunta deben crear una tabla dinámica utilizando las variables que corresponden
para abordarla y contrastarlos con algún criterio definido por ustedes que dé cuenta de la
“mayoría de los aranceles”.
En esta pregunta resultan fundamentales 3 puntos:
1) Creación correcta de la tabla dinámica y gráfico apropiado.
2) Cálculo de las medidas de posición a partir del criterio establecido para la variable
respectiva.
3) Análisis y conclusión en torno a la información obtenida.

```python
import pandas as pd
import numpy as np

# medidas de posicion y outliers

# cuariles
Q1 = df_1['VALOR ARANCEL (PESOS)'].quantile(0.25)
Q3 = df_1['VALOR ARANCEL (PESOS)'].quantile(0.75)

# rango intercualitativo
IQR = Q3 - Q1

# limite superior del outlier
limite_outlier = Q3 + 1.5 * IQR

print(f"Cuartil 1 (Q1 - 25% más barato): {Q1:,.0f} PESOS")
print(f"Cuartil 3 (Q3 - 25% más caro): {Q3:,.0f} PESOS")
print(f"Límite Superior para Outliers (Q3 + 1.5*IQR): {limite_outlier:,.0f} PESOS")


# creacion de la tabla dinamica de outliers

# filtrar el dataframe para obtener solo los datos que superen el outlier
carreras_outliers = df_1[df_1['VALOR ARANCEL (PESOS)'] > limite_outlier].copy()

# tabla dinamica con las carreras mas caras
# agrupar para mostrar el valor del arancel y cuantas veces aparece
tabla_outliers = carreras_outliers.groupby(['NOMBRE CARRERA', 'AREA CONOCIMIENTO'])['VALOR ARANCEL (PESOS)'].agg(
    Arancel_Unico = 'first',
    N_Veces_Registrada = 'count'
).sort_values(by='Arancel_Unico', ascending=False).reset_index()

# nombre columnas
tabla_outliers.columns = [
    'Carrera',
    'Área del Conocimiento',
    'Valor Arancel (PESOS)',
    'N° Registros'
]

print("\n--- TABLA DE CARRERAS CON ARANCEL SUSTANTIVAMENTE MÁS CARO (OUTLIERS) ---")
print(tabla_outliers.head(10)) # 10 carreras mas cras
```

```python
# --- diagrama de caja y bigotes--

plt.figure(figsize=(10, 6))

# boxplot para graficar distribucion de aranceles
plt.boxplot(
    df_1['VALOR ARANCEL (PESOS)'].dropna(), # NaNs
    vert=False # grafico horizontal
)

# titulo
plt.title('Distribución del Valor Arancel y Carreras Outliers', fontsize=16)
plt.xlabel('Valor Arancel (PESOS)', fontsize=12)
plt.yticks([1], ['Aranceles']) # Etiqueta el eje Y
plt.grid(axis='x', linestyle='--', alpha=0.7)

#marcar el limite del outlier
plt.axvline(limite_outlier, color='red', linestyle='--', label=f'Límite Outlier ({limite_outlier:,.0f})')
plt.legend()
plt.show()
```
